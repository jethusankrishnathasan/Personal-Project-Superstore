```python
# Import Pandas for data cleaning
import pandas as pd

# Mount Google Drive to access the dataset
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Load the dataset from drive into the Pandas DataFrame (used latin1 encoding to avoid decoding errors)
dataframe = pd.read_csv('/content/drive/MyDrive/Superstore_Raw.csv', encoding='latin1')
```

```python
# Check for duplicate rows
dataframe.duplicated().sum()
```

```python
# Convert date columns to proper datetime
dataframe['Order Date'] = pd.to_datetime(dataframe['Order Date'], errors='coerce')
dataframe['Ship Date'] = pd.to_datetime(dataframe['Ship Date'], errors='coerce')
```

```python
# Convert numeric columns
dataframe['Sales'] = pd.to_numeric(dataframe['Sales'], errors='coerce')
dataframe['Quantity'] = pd.to_numeric(dataframe['Quantity'], errors='coerce')
dataframe['Discount'] = pd.to_numeric(dataframe['Discount'], errors='coerce')
dataframe['Profit'] = pd.to_numeric(dataframe['Profit'], errors='coerce')
```

```python
import csv
# Export the cleaned dataset for the analysis
dataframe.to_csv('Superstore_Cleaned.csv',
                 sep=',',
                  index=False,
                  encoding='latin1',
                  quoting=csv.QUOTE_ALL
)
```
